## Write a Function to Scrape all of the Events on the Given Page Events Page

The function should return a Pandas DataFrame with columns for the Event_Name, Venue, Event_Date and Number_of_Attendees.

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('data.sqlite')
c = conn.cursor()
```

```python
def scrape_events(events_page_url):
    soup = BeautifulSoup(html_page.content, 'html.parser')
    event_name = soup.findAll('td', {'class':'overview-top'})
    dates = soup.findAll('div', {'class':'list detail sub-list'})
    dicts = []
    for event in event_name:
        info = {'Movie_Name': event.h4.a['title'], 
                'Meta_score': event.div.findAll('span', {'class':'metascore'})[0].text.strip() if bool(event.div.select('span', {'class':'metascore'})) == True else None,
                'Director': event.findAll('div',{'class':'txt-block'})[0].span.a.text}
        dicts.append(info)
    return dicts
```

```python
movies_dict = scrape_events(html_page)
movies_dict
```

```python
c.execute("""DROP TABLE movies""")
```

```python
c.execute("""CREATE TABLE movies (id INTEGER PRIMARY KEY AUTOINCREMENT, Movie_Name TEXT , Meta_Score INTEGER, Director TEXT)""")
```

```python
def populate_sql(dicts):
    movie_name = str(dicts['Movie_Name'])
    meta_score = str(dicts['Meta_score'])
    director = str(dicts['Director'])
    
    c.execute("""INSERT INTO movies
                 (Movie_Name, Meta_Score, Director)
                 VALUES(?, ?, ?)""", (movie_name, meta_score, director))
for dic in movies_dict:
    populate_sql(dic)
        
```
